```python
from matplotlib import pyplot as plt
import numpy as np
import cv2 as cv
```

```python
image = cv.imread("server.png")

orig_size = image.shape
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
#image = image[::2,::2,:]
#print(image.shape,(image.shape[0]//2, image.shape[1]//2))
smallimage = cv.resize(image, (image.shape[1]//2, image.shape[0]//2))
print(smallimage.shape)
print(orig_size)
plt.imshow(smallimage)
plt.show()
```

```python
image_NN = cv.resize(smallimage, (orig_size[1], orig_size[0]), interpolation = cv.INTER_NEAREST)
image_Linear = cv.resize(smallimage, (orig_size[1], orig_size[0]), interpolation = cv.INTER_LINEAR)
image_Cubic = cv.resize(smallimage, (orig_size[1], orig_size[0]), interpolation = cv.INTER_CUBIC)

plt.imshow(image_NN)
plt.title("Nearest Neighbour")
plt.show()

plt.imshow(image_Linear)
plt.title("Bilinear")
plt.show()

plt.imshow(image_Cubic)
plt.title("Bicubic")
plt.show()
```

```python
#Task 2
Gafilt= np.array([[1,2,1], [2,4,2], [1,2,1]])/16
Gaussian = cv.filter2D(image, 3, Gafilt)
plt.imshow(Gaussian)
plt.title("Gaussian Filter Applied")
plt.show()

Sharp = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
Sharpened = cv.filter2D(image, 3, Sharp)
plt.imshow(Sharpened)
plt.title("Sharpening Applied")
plt.show()

Median = cv.medianBlur(image, 3)
plt.imshow(Median)
plt.title("Median Applied")
plt.show()

```
