```python
from matplotlib import pyplot as plt
import numpy as np
import cv2 as cv
```

```python
image = cv.imread("Fig01.tif")
#image = cv.cvtColor(image, cv.COLOR_BGR2RGB)/255
plt.imshow(image)
plt.show()

image3 = cv.imread("Fig03.tif")
#image3 = cv.cvtColor(image3, cv.COLOR_BGR2RGB)/255
plt.imshow(image3)
plt.show()
```

```python
#Task 1
Avgfilt = np.array([[1,1,1], [1,1,1], [1,1,1]])/9
Averaged = cv.filter2D(image, 3, Avgfilt)
plt.imshow(Averaged)
plt.title("Averaging Filter Applied")
plt.show()

Gafilt= np.array([[1,2,1], [2,4,2], [1,2,1]])/16
Gaussian = cv.filter2D(image, 3, Gafilt)
plt.imshow(Gaussian)
plt.title("Gaussian Filter Applied")
plt.show()
```

```python
#Task 2
Median3 = cv.medianBlur(image, 3)
Median15 = cv.medianBlur(image, 15)
Median31 = cv.medianBlur(image, 31)

plt.imshow(Median3)
plt.title("Median 3x3 Filter Applied")
plt.show()
plt.imshow(Median15)
plt.title("Median 15x15 Filter Applied")
plt.show()
plt.imshow(Median31)
plt.title("Median 31x31 Filter Applied")
plt.show()


minmaxFilt3 = np.ones((3,3))
minmaxFilt15 = np.ones((15,15))
minmaxFilt31 = np.ones((31,31))


Min3 = cv.erode(image, minmaxFilt3)
Min15 = cv.erode(image, minmaxFilt15)
Min31 = cv.erode(image, minmaxFilt31)

plt.imshow(Min3)
plt.title("Min 3x3 Filter Applied")
plt.show()
plt.imshow(Min15)
plt.title("Min 15x15 Filter Applied")
plt.show()
plt.imshow(Min31)
plt.title("Min 31x31 Filter Applied")
plt.show()

Max3 = cv.dilate(image, minmaxFilt3)
Max15 = cv.dilate(image, minmaxFilt15)
Max31 = cv.dilate(image, minmaxFilt31)

plt.imshow(Max3)
plt.title("Max 3x3 Filter Applied")
plt.show()
plt.imshow(Max15)
plt.title("Max 15x15 Filter Applied")
plt.show()
```

```python
#Task 3
Horz_Sobel = np.array([[-1,-2,-1], [0,0,0], [1,2,1]])
Vert_Sobel = np.array([[-1,0,1], [-2,0,2], [-1,0,1]])

Horz = cv.filter2D(image, 3, Horz_Sobel)
Vert = cv.filter2D(image, 3, Vert_Sobel)

plt.imshow(Horz)
plt.title("Horizontal Sobel Filter Applied")
plt.show()
plt.imshow(Vert)
plt.title("Vertical Sobel Filter Applied")
plt.show()

Edges = (cv.normalize(np.sqrt(np.square(np.int32(Horz)) + np.square(np.int32(Vert))), None, alpha=0, beta=255, norm_type=cv.NORM_MINMAX)).astype(int)
plt.imshow(Edges)
plt.title("Combined Edges")
plt.show()
```

```python
#Task 4
Lapfilt = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]])
Laplacian = cv.filter2D(image3, 3, Lapfilt)

Lapfilt2 = np.array([[0,-1,0], [-1,4,-1], [0,-1,0]])
Laplacian2 = cv.filter2D(image3, 3, Lapfilt2)

Sharpened1 = image3 + Laplacian
Sharpened2 = image3 + Laplacian2

plt.imshow(Laplacian)
plt.title("Laplacian Filter Applied")
plt.show()
plt.imshow(Sharpened1)
plt.title("Sharpened Image")
plt.show()

plt.imshow(Laplacian2)
plt.title("2nd Laplacian Filter Applied")
plt.show()
plt.imshow(Sharpened2)
plt.title("2nd Filter Sharpened Image")
plt.show()

#Extra Task
Lapfilt = np.array([[1,1,1], [1,-8,1], [1,1,1]])
Laplacian = cv.filter2D(image3, 3, Lapfilt)

Lapfilt2 = np.array([[0,1,0], [1,-4,1], [0,1,0]])
Laplacian2 = cv.filter2D(image3, 3, Lapfilt2)

Sharpened = image3 - Laplacian
Sharpened2 = image3 - Laplacian2

plt.imshow(Laplacian)
plt.title("Laplacian Filter Applied")
plt.show()
plt.imshow(Sharpened)
plt.title("Sharpened Image (Subtraction)")
plt.show()

plt.imshow(Laplacian2)
plt.title("2nd Laplacian Filter Applied")
plt.show()
plt.imshow(Sharpened2)
plt.title("2nd Filter Sharpened Image (Subtraction)")
plt.show()
```
